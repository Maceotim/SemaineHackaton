```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dg =df.df[['P2','Lt', 'P4','P5','V','E2','P6_reel','E2','S_moy','sigma_S','Lq_reel']]
dg= dg.dropna()
```

```python
X = np.array(dg[['P2','Lt', 'P4','P5','V','E2','P6_reel','E2','Lq_reel']])
Y= np.array(dg[['S_moy', 'sigma_S']])
```

```python
from sklearn import model_selection
from sklearn import preprocessing
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, 
                                                                    test_size=0.20, 
                                                                    random_state=17)

print(len(X_train))
```

```python
x_grid = np.linspace(0, 1, 100)
x_grid_2d = x_grid.reshape((100, 1))
# Instancier un objet permettant de créer des variables polynomiales de degré au plus 15
polynomial_features = preprocessing.PolynomialFeatures(degree=6, include_bias=False)

# Appliquer cet objet aux variables de X_train 
X_train_poly = polynomial_features.fit_transform(X_train)

# Appliquer la même transformation aux données de test
X_test_poly = polynomial_features.transform(X_test)

from sklearn import linear_model
polyreg_ridge=linear_model.Ridge(alpha=0.05)
polyreg_ridge.fit(X_train_poly, Y_train)
Y_pred = polyreg_ridge.predict(X_train_poly)

from sklearn import metrics
print(f"RMSE: {metrics.root_mean_squared_error(Y_train, Y_pred):.2f}")
print(f"RMSE: {metrics.root_mean_squared_error(Y_test, polyreg_ridge.predict(X_test_poly) ):.2f}")
```
