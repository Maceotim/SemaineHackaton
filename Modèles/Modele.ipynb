```python
import pandas as pd
import numpy as np

data = pd.ExcelFile("data_CRIR_anon.xlsx")

data_P = pd.read_excel(data, 'P') #Ouverture des différentes tables
data_S = pd.read_excel(data, 'S')
data_EF = pd.read_excel(data, 'EF')
data_T = pd.read_excel(data, 'T')

##on enlève la première ligne qui correspond à des commentaires
data_P.drop(0, inplace=True)
data_S.drop(0, inplace=True)
```

```python
data_EF_clean = data_EF.drop(data_EF[data_EF['P6_reel']==1.00].index)
data_EF_clean.drop(0, inplace=True)
data_EF_clean.head(10)
```

```python
data_EF_clean['round_P6']= data_EF_clean['P6_reel'].round(0)
```

```python
df= pd.merge(data_EF_clean, data_P, how='left', left_on=['essai','round_P6'] , right_on=['essai','P6_target'])
```

```python
data_S['S1'] = pd.to_numeric(data_S['S1'], errors='coerce')
data_S_m = data_S.groupby('essai', as_index=False).agg(
    S_moy=('S1', 'mean'),
    sigma_S=('S1', 'std')
)
data_S_m['essai'] = data_S_m['essai'].astype(int)
data_S_m.head()
```

```python
df= pd.merge(df, data_S_m, how='left', left_on=['essai'] , right_on=['essai'])
df.to_csv('mes_donnees.csv', sep=';', index=False, encoding='utf-8-sig')
df.head(25)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_normalize = df.copy() #copie du tableau
df_normalize_R = df_normalize[df_normalize['Lt'] == 'R'] #On sépare les données en deux tableaux, un pour le Lt=R et un pour le Lt=G
df_normalize_G = df_normalize[df_normalize['Lt'] == 'G']
df_normalize_R['Lt'], df_normalize_G['Lt'] = df_normalize_R['Lt'].map({'R': 0}), df_normalize_G['Lt'].map({'G': 1}) #On remplace les valeurs de Lt par des nombres

df_normalize_R['F1'], df_normalize_G['F1'] = df_normalize_R['F1'].map({'Cold': 0, 'FME': 1}), df_normalize_G['F1'].map({'Cold': 0, 'FME': 1})
df_normalize['F1'] = df_normalize['F1'].map({'Cold': 0, 'FME': 1})
df_normalize['Lt'] = df_normalize['Lt'].map({'R': 0, 'G': 1})
colonnes_a_enlever = ['date', 'condition', 'essai', 'P6_reel'] #On enlève les colonnes qui ne sont pas des nombres et qui ne sont pas utiles
masque = ~df_normalize.columns.isin(colonnes_a_enlever)

df_normalize.loc[:, masque] = scaler.fit_transform(df_normalize.loc[:, masque]) #On normalise
df_normalize_R.loc[:, masque] = scaler.fit_transform(df_normalize_R.loc[:, masque]) #On normalise
df_normalize_G.loc[:, masque] = scaler.fit_transform(df_normalize_G.loc[:, masque]) #On normalise

df_normalize['E2.1.3_moy'] = df_normalize[['E2.1', 'E2.3']].mean(axis=1) #On fait la moyenne de E2.1 et E2.3
df_normalize['E2.2.4_moy'] = df_normalize[['E2.2', 'E2.4']].mean(axis=1) #On fait la moyenne de E2.2 et E2.4
df_normalize['E2_moy'] = df_normalize[['E2.1','E2.2','E2.3','E2.4']].mean(axis=1) #On fait la moyenne de E2.1, E2.2, E2.3 et E2.4

df_normalize_R['E2.1.3_moy'] = df_normalize_R[['E2.1', 'E2.3']].mean(axis=1) #On fait la moyenne de E2.1 et E2.3
df_normalize_R['E2.2.4_moy'] = df_normalize_R[['E2.2', 'E2.4']].mean(axis=1) #On fait la moyenne de E2.2 et E2.4
df_normalize_R['E2_moy'] = df_normalize_R[['E2.1','E2.2','E2.3','E2.4']].mean(axis=1) #On fait la moyenne de E2.1, E2.2, E2.3 et E2.
```

```python
df_normalize.to_csv('mes_donnees_normalisées.csv', sep=';', index=False, encoding='utf-8-sig') #exportation des données normalisées en csv
df_normalize_R.to_csv('mes_donnees_normalisées_R.csv', sep=';', index=False, encoding='utf-8-sig') #exportation des données normalisées en csv
df_normalize_G.to_csv('mes_donnees_normalisées_G.csv', sep=';', index=False, encoding='utf-8-sig') #exportation des données normalisées en csv
```
